```python
# Load libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split # Import train_test_split function
import numpy as np
import gensim
from scipy.sparse import hstack
import utils


model = gensim.models.word2vec.Word2Vec.load("updated_google_corpus")

TESTING_MODE = False
```

```python
# load dataset
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
test_data = test_data.drop("index",axis=1)
data = pd.concat([train_data,test_data],sort=False)
```

```python
d = pd.get_dummies(train_data[["country","region_2","province","taster_name","variety"]])
d["points"]= train_data["points"]
d["id"] = train_data['id']
```

```python
description_vecs = train_data["description"].apply(lambda words:sum([model[word.lower()] for word in words.split()])*1/len(words.split()))

for i in range(300):
    d["vec_dim_"+str(i)]=description_vecs.apply(lambda v:v[i])
```

```python
X_train,y_train,X_test, y_test = d[:175000], data["price"][:175000], d[175000:], data["price"][175000:]
```

```python
X_train, X_test, y_train, y_test = train_test_split(d, train_data["price"], test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
X_train = X_train.drop(["id"],axis=1)
```

```python
%%time
# Create Decision Tree classifer object
clf = LinearRegression(n_jobs=10)
# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)
```

```python
#Predict the response for test dataset
ids = X_test["id"]
X_test = X_test.drop(["id"],axis=1)
```

```python
y_pred = clf.predict(X_test)
```

```python
# removing outliers
y_pred[y_pred>2500] = 50
y_pred[y_pred<0]=1
```
