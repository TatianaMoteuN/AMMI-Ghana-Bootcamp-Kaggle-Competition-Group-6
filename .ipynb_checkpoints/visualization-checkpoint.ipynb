```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
import numpy as np
import xgboost
import gensim
from scipy.sparse import hstack

wv = gensim.models.word2vec.Word2Vec.load("data/updated_google_corpus")
```

```python
# load dataset
train_data = pd.read_csv("data/train.csv")
test_data = pd.read_csv("data/test.csv")
test_data = test_data.drop("index",axis=1)
data = pd.concat([train_data,test_data],sort=False)
```

```python
train_data.head()
```

```python
def vectorise(sentance):
    vecs = []
    for word in sentance.lower().split():
        try:
            vecs.append(wv[word])
        except Exception as e:
            print(e)
            vecs.append(np.zeros((300,1)))
    return sum(vecs)
```

```python
d2 = pd.DataFrame()
if TESTING_MODE:
    for col in ["description"]:
        vecs = train_data[col].apply(lambda words:vectorise(words))
        for i in range(300):
            d2["vec_"+col+"_dim_"+str(i)]=vecs.apply(lambda v:v[i])
else:
    for col in ["description"]:
        vecs = data[col].apply(lambda words:vectorise(words))
        for i in range(300):
            d2["vec_"+col+"_dim_"+str(i)]=vecs.apply(lambda v:v[i])
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
principalComponents = pca.fit_transform(d2)
```

```python
d = pd.DataFrame(data = principalComponents,columns=["pca"+str(i) for i in range(10)])
```
